```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_points_x = 10
num_points_y = 10

x_coords = np.linspace(0, 1, num_points_x)
y_coords = np.linspace(0, 1, num_points_y)

grid = np.zeros((100,2))

for i in range(len(x_coords)):
    for j in range(len(y_coords)):
        grid[10*i+j,0] = x_coords[i]
        grid[10*i+j,1] = y_coords[j]
```

```python
plt.scatter(grid[:,0],grid[:,1],color='black')
```

```python
class_ = np.random.permutation(np.arange(100))
class_A = class_[:10]
class_B = class_[10:20]
class_val = class_[20:]
```

```python
class_B
```

```python
class_val
```

```python
target = np.zeros((100))
target[class_A] = 1
target[class_B] = -1
```

```python
target
```

```python
training_set = np.append(class_A,class_B)
print(training_set)
```

```python
plt.scatter(grid[:,0],grid[:,1],color='black')
plt.scatter(grid[class_A,0],grid[class_A,1],color='red')
plt.scatter(grid[class_B,0],grid[class_B,1],color='blue')
```

```python
from scipy.spatial import Voronoi,voronoi_plot_2d
```

```python
vor = Voronoi(grid[class_A])
vor2 = Voronoi(grid[class_B])
voronoi_plot_2d(vor)
voronoi_plot_2d(vor2)
plt.show()
```

```python
dist = np.zeros((grid.shape[0],grid.shape[0]))
for i in range(grid.shape[0]):
    for j in range(grid.shape[0]):
        point_1 = grid[i,:]
        point_2 = grid[j,:]
        dist[i,j] = np.linalg.norm(point_1 - point_2)
```

```python
class_val
```

```python
dist[20,training_set]
```

```python
np.argsort(dist[20,training_set])
```

```python
for i in class_val:
    if(np.sort(dist[i,training_set])[0] != np.sort(dist[i,training_set])[1]):
        idx = np.argsort(dist[i,training_set])[0]
        nearest_neighbour = training_set[idx]
        target[i] = target[nearest_neighbour]
        if(target[i]==1):
            class_A = np.append(class_A , i)
            
        if(target[i]==-1):
            class_B = np.append(class_B , i)
            
    
```

```python
distance = np.zeros((len(training_set)))
for i in range(len(training_set)):
    point_1 = point
    point_2 = grid[training_set[i],:]
    distance[i] = np.linalg.norm(point_1 - point_2)
    
nearest_point = np.argsort(distance)[0]
class_predicted = target[training_set[nearest_point]]

if(class_predicted == 1):
    print("point belongs to red class")
if(class_predicted == -1):
    print("point belongs to blue class")
```
