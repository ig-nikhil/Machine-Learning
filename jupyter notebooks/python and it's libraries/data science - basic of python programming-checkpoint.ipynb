# function in python

```python
def findMin(L):
    m = L[0]
    idx = 0
    i = 0
    for x in L:
        if x<m:
            m = x
            idx = i
        else:
            pass
        i+=1
    return m, idx

```

```python
a,b = findMin([2,3,4,1,9])
print(a,b)
```

here in python language we can return more than one value using comma as in above function we return m and idx simultaneously.
