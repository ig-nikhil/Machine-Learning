```python
plt.style.use("dark_background")
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos (x)
y3 = y1+y2 
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
fig = plt.figure(figsize = (4,3))
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
fig = plt.figure(figsize = (4,3))
ax = plt.axes()
ax.plot(x,y1,linewidth = 5)
ax.plot(x,y2)
```

```python
fig = plt.figure(figsize = (6,5),dpi = 90)
plt.plot(x,y1,'-r')
# plt.xlim(-1,12)
# plt.ylim(-3,3)
plt.axis([-1,12,-3,3]) # anohter way to setting x and y limits
#plt.axis('tight')
```

```python
plt.plot(x,x,'-k')
plt.xlim(10,0) # reversing the limit leads to flip the plot
```

# Labels and Legends

```python
fig = plt.figure(dpi=100)
plt.style.use("dark_background")
plt.plot(x,y1,label = 'sin(x)')
plt.plot(x,y2,label = 'cos(x)')
plt.plot(x,y3,label = 'sin(x)+cos(x)')
plt.legend(loc = 'lower left')
plt.title('sin-cos-curve')
plt.xlabel('x values')
plt.ylabel('y values')
```

```python
markers = ['o','+','^','v','>','<','s','d'] # markers in the python
fig = plt.figure(figsize = (4,3),dpi = 100)
X = np.linspace(0,10,30)
Y = np.sin(X)
plt.plot(X,Y,markers[6],color='white')
```

# Countour Plot

```python
fig = plt.figure()
X,Y = np.meshgrid(np.linspace(5,10,50),np.linspace(5,10,80))
Z = X+(3*Y)*9*X
plt.contourf(X,Y,Z,40, cmap = 'inferno')
plt.colorbar()
```
