## dataframe visuilzation using matplotlib and seaborn

```python
X = pd.Series (50*np.random.randn(5000) )
Y = pd.Series (200*np.random.randn(5000))
Z = pd.Series (100*np.random.randn(5000) + 500)
plt.hist(X , alpha = 0.5 , density = True)
plt.hist(Y , alpha = 0.5 , density = True)
plt.hist(Z , alpha = 0.5 , density = True)

```

```python
data= pd.DataFrame({'X' : X, 'Y' : Y, 'Z' : Z})
for col in data.columns:
    plt.hist(data[col], alpha = 0.5, density = True)
```

```python
data= pd.DataFrame({'X' : X, 'Y' : Y, 'Z' : Z})
for col in data.columns:
    plt.hist(data[col], alpha = 0.5, density = True)
    sns.kdeplot(data[col],shade=True)
```

```python
for col in data.columns:
    sns.distplot(data[col])
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_linearly_separable_data(size=100, seed=None):
    if seed is not None:
        np.random.seed(seed)

    # Generate random weights for the linear decision boundary
    weights = np.random.rand(2)

    # Generate random data points for two classes
    class1 = np.random.randn(size, 2) + 2
    class2 = np.random.randn(size, 2) - 2

    # Create labels for the two classes
    labels_class1 = np.ones(size)
    labels_class2 = -np.ones(size)

    # Combine the data and labels
    data = np.vstack((class1, class2))
    labels = np.hstack((labels_class1, labels_class2))

    return data, labels, weights

# Generate linearly separable data
size = 100
data, labels, weights = generate_linearly_separable_data(size=size, seed=42)

# Plot the data points
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap=plt.cm.Paired)
plt.title('Linearly Separable Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot the decision boundary
x_range = np.linspace(min(data[:, 0]), max(data[:, 0]), 100)
decision_boundary = -(weights[0] / weights[1]) * x_range
plt.plot(x_range, decision_boundary, color='black', linestyle='dashed', linewidth=2)

plt.show()
```
