```python
import numpy as np 
import matplotlib.pyplot as plt 
```

### Generating linearly separable data

```python
# y = 0.5x + 4
x = np.random.randint(1,20,size=(20))
y = np.random.randint(1,20,size=(20))
print(x)
print(y)
class_1 = np.zeros((20,2))
class_2 = np.zeros((20,2))
```

```python
j=0
k=0
for i in range(20):
    if ( (y[i] - (0.5)*x[i] - 4) > 0):
        class_1[j][0] = x[i]
        class_1[j][1] = y[i]
        j=j+1
    else:
        class_2[j][0] = x[i]
        class_2[j][1] = y[i]
        j=j+1
        
```

```python
class_1
```

```python
class_2
```

```python
plt.scatter(class_1[:,0],class_1[:,1],color="g")
plt.scatter(class_2[:,0],class_2[:,1],color="b")
plt.plot()
plt.title('Linearly Separable Data')
plt.xlabel('Feature 1 ,x2')
plt.ylabel('Feature 2 ,x3')
p = np.linspace(-2,25,1000)
plt.plot(p,((0.5)*p +4),'-c')
plt.legend( 'AB', ncol=2, loc='upper left')
```

```python
class_1 = class_1[~np.all(class_1 == 0,axis=1)]
print(class_1)
```

```python
class_2 = class_2[~np.all(class_2 == 0,axis=1)]
print(class_2)
```

## perceptron learning algorithm implementation

```python
class_i = np.vstack((class_1 , class_2))
class_i
```

```python
idx = np.random.permutation(np.arange(20))
class_3 = np.zeros((20,2))
Y_3 = np.ones((20,1))
j=0
for i in idx :
    if(i >= 10):
        class_3[j] = class_i[i]
        Y_3[j] = -1
        j = j+1
    else:
        class_3[j] = class_i[i]
        j = j+1

print(Y_3)
print(class_3)
```

```python
X = np.hstack((np.ones((20,1)),class_3))
X
```

## perceptron_algorithm


```python
import matplotlib.pyplot as plt
import numpy as np

w = [1, -4, 1]
training_set = X
result = Y_3
test_set_class1 = class_1
test_set_class2 = class_2

def plot_decision_boundary(w, title):
  
    plt.scatter(test_set_class1[:, 0], test_set_class1[:, 1], color='yellow', label='Class 1')
    plt.scatter(test_set_class2[:, 0], test_set_class2[:, 1], color='green', label='Class 2')

    x_line = np.linspace(0, 100, 100)
    y_line = (-w[1] * x_line - w[0]) / w[2]
    plt.plot(x_line, y_line, color='red', label='Decision Line')

    plt.title(title)
    plt.legend()

    
    plt.xlim(0, 30)
    plt.ylim(0, 30)

    plt.show(block=False)
no_of_updates = 0
for i in range(100):
    for j in range(len(training_set)):
        Z = np.dot(w, training_set[j])

        if np.sign(Z) != result[j]:
            w = w + result[j] * training_set[j]
            no_of_updates = no_of_updates + 1

    plot_decision_boundary(w, f"Perceptron learning - Iteration {i + 1}")

   

 
print("No of Updates perform by this algoritm before converging is :",no_of_updates)
print("value of w is :",w)
```
