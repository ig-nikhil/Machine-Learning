```python
X = np.random.randint(0,10,size=(4,2))
S = np.dot(X,(X.T))
is_symmetric = np.allclose(S, S.T)

if is_symmetric:
    print("The matrix is symmetric.")
else:
    print("The matrix is not symmetric.")
```

```python
eigen_values = la.eigvals(S)
is_positive_semi_definite = all(eig >= 0 for eig in eigen_values)

if is_positive_semi_definite:
    print("The matrix is positive semi-definite.")
else:
    print("The matrix is not positive semi-definite.")

is_positive_definite = all(eig > 0 for eig in eigen_values)

if is_positive_definite:
    print("The matrix is positive definite.")
else:
    print("The matrix is not positive definite.")    
```

```python
A = np.array([[2,1,3],[1,1,2],[3,2,5]])
```

```python
la.eigvals(A)
```

```python
det_A = la.det(A)
trace_A = np.trace(A)

# Solve the characteristic equation: det(A - λI) = 0
# The characteristic polynomial is λ^3 - trace(A)λ^2 + (trace(A)^2 - det(A))λ - (trace(A)^2 - det(A))det(A) = 0
# Solve for λ using numpy.roots
eigenvalues = np.roots([1, -trace_A, (trace_A**2 - det_A), -(trace_A**2 - det_A)*det_A])
print("Eigenvalues:", eigenvalues)
```

```python
eigenvalues, eigenvectors = la.eig(A)
```

```python
# Reconstruct the original matrix using eigenvalues and eigenvectors
A_reconstructed = np.dot(np.dot(eigenvectors, np.diag(eigenvalues)), np.linalg.inv(eigenvectors))

print("Reconstructed matrix:\n", A_reconstructed)
```
