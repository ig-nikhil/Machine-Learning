```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
# X = no of hours of study by student , Y = rank of the student

no_of_sample = 30
X = (np.random.rand(no_of_sample,1))*24
X = np.sort(X , axis=0)

Y = np.arange(1,no_of_sample+1)
Y = Y[::-1].reshape(no_of_sample,1)
```

```python
# plot of X and Y 

plt.scatter(X,Y)
```

```python
W = np.ones((2,1))
ephocs = 100
lr = 0.001
```

```python
for i in range(ephocs):
    for i in range(X.shape[0]):
        
        idx = np.random.randint(0,X.shape[0])
        
        y_predicted = W[0][0] + W[1][0]*X[idx]
        
        
        slope_W_0 = -2*(Y[idx]-y_predicted) 
        W[0][0] = W[0][0] - lr*slope_W_0
        
        slope_W_1 = -2*(Y[idx]-y_predicted)*X[idx]
        W[1][0] = W[1][0] - lr*slope_W_1
        
        x_values = np.linspace(0, 25, 200)
        y_values = W[0][0] + x_values*W[1][0]
        plt.scatter(X,Y)
        plt.plot(x_values,y_values)
        
print(W)       
```
